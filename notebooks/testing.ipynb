```python


PROJECT_ROOT = Path(r"C:\Users\zxcvk\Desktop\personal\thamanyah\ml-churn-project")   
DATA_PATH = PROJECT_ROOT / "data" / "customer_churn_mini.json"

def load_data(path: Path) -> pd.DataFrame:
    """
    Load the customer churn data from a JSON file and return a pandas DataFrame.
    """
   
    try:
        df = pd.read_json(path, lines=True)
    except ValueError:
        df = pd.read_json(path)

    return df
```

```python
print(f"Loading data from: {DATA_PATH}")

df = load_data(DATA_PATH)
```
